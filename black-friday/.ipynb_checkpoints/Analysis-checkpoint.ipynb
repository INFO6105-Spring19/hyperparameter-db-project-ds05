# Analysis: Finding the Range of Hyperparameters

# Abstract

We are generating the range of the hyperparameters generated in my dataset for every hyperparameter family. In this Jupyter notebook we are collecting the hyperparameters for the run time 500, 1000, 1500, 2000, 2500 Seconds. and by doing analysis will find the range of the hyperparameters

```python
RatingL = df['Col_Sample_Rate'].unique()
RatingDict = {}
for i in range(len(RatingL)):
    RatingDict[RatingL[i]] = i
df['Col_Sample_Rate'] = df['Col_Sample_Rate '].map(RatingDict).astype(float)


RatingL = df['Max_abs_leafnode_pred'].unique()
RatingDict = {}
for i in range(len(RatingL)):
    RatingDict[RatingL[i]] = i
df['Max_abs_leafnode_pred'] = df['Max_abs_leafnode_pred'].map(RatingDict).astype(float)


RatingL = df['Distribution'].unique()
RatingDict = {}
for i in range(len(RatingL)):
    RatingDict[RatingL[i]] = i
df['Distribution'] = df['Distribution'].map(RatingDict).astype(float)


RatingL = df['Max_abs_leafnode_pred'].unique()
RatingDict = {}
for i in range(len(RatingL)):
    RatingDict[RatingL[i]] = i
df['Max_abs_leafnode_pred'] = df['Max_abs_leafnode_pred'].map(RatingDict).astype(float)


```
