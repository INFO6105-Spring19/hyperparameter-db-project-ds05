# Analysis: Finding the Range of Hyperparameters

# Abstract

We are generating the range of the hyperparameters generated in my dataset for every hyperparameter family. In this Jupyter notebook we are collecting the hyperparameters for the run time 500, 1000, 1500, 2000, 2500 Seconds. and by doing analysis will find the range of the hyperparameters

```python
df= df.drop(columns=['Max_abs_leafnode_pred'])
df
```

```python
X= df.iloc[:,1:]
X
```

```python
ranges = []
for key,val in df.dtypes.items():
    print(key,val)
    if val == 'int': 
         ranges.append([key,[df[key].min(),df[key].max()]])
    if val == 'float':
         ranges.append([key,[df[key].min(),df[key].max()]])
```
